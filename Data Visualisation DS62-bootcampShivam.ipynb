## Boxplot

#### It is used to see the quartile wise distibution of numerical column
#### It is used to check the outliers
#### It is used to see the quartile wise relatioship between categrorical and numerical data

```python
df["temp"].describe()
```

```python
sns.boxplot(x=df["temp"])
plt.show()
#whiskers/handles
#quartile = Quarter + Percentile (25%,50%,75%)
# First (leftmost) side of box --> First Quartile (Q1-25%)
# Middle (Center) side of box --> Second Quartile (Q2-50%) (Median)
# Last (Rightmost) side of box --> Third Quartile (Q3-75%)
# Emperical Relationships
# Lower Whiskers = Q1-1.5*(Q3-Q1)
# Higher Whiskers = Q3+1.5*(Q3-Q1)
```

```python
sns.boxplot(x=df["windspeed"])
plt.show()
```

```python
sns.boxplot(x=df["hum"])
plt.show()
```

```python
sns.boxplot(x=df["mnth"],y=df["temp"])
plt.show()
#x=categorical
#y=numerical
```

```python
sns.boxplot(y=df[df["mnth"]==9## Scatterplot 
#### Relation of two numerical variable]["temp"])
plt.show()
```

## Scatterplot 
#### Relationship of two numerical variable

```python
sns.scatterplot(x=df["temp"],y=df["cnt"])
plt.show()
#Positive (if one variable increases other variable also increases)
#Negative (if one variable increases other variable decreases)
#No Relation (No relationship)
```

```python
sns.scatterplot(x=df["temp"],y=df["windspeed"])
plt.show()
```

```python
sns.scatterplot(x=df["temp"],y=df["atemp"])
plt.show()
```

## PairPlot

#### It is the combination of histplot+scatterplot

```python
sns.pairplot(df[["temp","hum","windspeed","cnt","atemp"]])
plt.show()
```

# HeatMap

```python
sns.heatmap(df[["temp","hum","windspeed","atemp"]].corr(),annot=True)
plt.show()
```
