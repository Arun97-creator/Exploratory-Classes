Using Pandas You Do

# Task : You are asked to find Player from 'Texas' College.

# Handling Multiple Conditions
---

```python
nba.head(3)
```

## Task : You are asked to find Players from 'Texas' College playing from 'PG' Position.
---

```python
c1 = nba.College == 'Texas'
c2 = nba.Position == 'PG'

nba[ c1 & c2 ]       # here & is Logical AND
```

```python
# direct

nba[ ( nba.College == 'Texas') & (nba.Position == 'PG') ]

# enclosing conditions in brackets is Mandatory
```

# Using OR Operator
---

# Task : You are asked to find players from either 'PG' or 'SF' Position.

```python
c1 = nba.Position == 'PG'
c2 = nba.Position == 'SF'

nba[ c1  | c2 ]     # here | is Logical OR
```

## Task : You are asked to find Players whose Salary is Missing (NaN)
---

```python
nba[ nba.Salary.isnull() ]
```

```python
nba[ nba.Salary.isna() ]
```

# Excluding the Rows where Salary Contains NaN
---

# ~ Negation
---

```python
nba[ ~ nba.Salary.isnull() ]
```
